###Importing libraries

```python
import numpy as np
import cv2
from PIL import Image
from IPython.display import display
```

###importing image to be converted

```python
img  =  cv2.imread("img.jpg")
img = cv2.resize(img,(640,1334))
rgb_img =  cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
display(Image.fromarray(rgb_img))
```

###Converting image to greyscale

```python
greyscale_img =  cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
display(Image.fromarray(greyscale_img))
```

###Converting image to negative

```python
neg_img = cv2.bitwise_not(greyscale_img)
display(Image.fromarray(neg_img))
```

```python
smooth_img = cv2.GaussianBlur(neg_img, (5,5), 0, 0)
display(Image.fromarray(smooth_img))
```

###Converting the negative to sketch

```python
sketch_img = 255 - smooth_img
sketch = cv2.divide(greyscale_img, sketch_img, scale=250.0)
display(Image.fromarray(sketch))
```

###Comparing the original image and the sketch

```python
rgb_img =  cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
display(Image.fromarray(rgb_img))
display(Image.fromarray(sketch))
```
